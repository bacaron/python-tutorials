## Module import
Software packages in python can be loaded into python as "modules". Loading a module is pretty simple, however there are a few specifics to note.

### Import entire module
Entire software packages can be loaded in a single line of code. In order to use these packages, they first must be loaded into the python environment. To load and manipulate the tract profile data later, we will want to load a few software packages. Specifically, pandas, numpy, matplotlib. Pandas is a software package that loads data as "dataframes", and is extremely useful for performing statistical analyses on .csv files (like the ones for tract profiles). In general, modules can be imported into the python environment by running the following command:

	import <software package name>

You can also specify the name of a module, usually to something shorter (we're all lazy), using this format:
	
	import <software package name> as <name you want to use>

```python
import numpy as np
import pandas as pd
```

For the purposes of this tutorial, we'll focus specifically on lists and dataframes.

### Indexing columns in the dataframe
A great deal of the manipulation done to dataframes requires indexing of either the columns or rows of data. With pandas, this is quite simple and can be quite powerful. Let's go over a few examples of indexing columns.

1. single column

Let's take a look at one column: 'structureID'. This field contains the track names for all of the 61 tracks segmented for each subject. Let's print this column:

```python
df['structureID']
```

We see a few rows including 'anteriorFrontalCC' and 'righttemporoThalamic'. How can we get all of the tract names? By using the 'unique' function. This function will only print out the unique names, and ignore the repeats of those name. Since there are 180 nodes x track, 61 tracks per subject, and 42 subjects, these names repeat quite frequently. The unique function can save us from the repeats. With pandas, we can do that by running the following:

```python
df['structureID'].unique()
```

This gives us an array (another datatype; we'll go over at a later date). However, it's much easier to work with in the form of a list using the tolist function:

```python
df['structureID'].unique().tolist()
```

Great! Let's store this list as a variable for later use:

```python
track_names = df['structureID'].unique().tolist()
track_names
```

Let's move onto indexing multiple columns!

2. indexing multiple columns

Indexing multiple columns is as easy as indexing a single column. Let's show this by indexing the 'subjectID' along with the 'structureID':

```python
df[['subjectID','structureID']]
```

See! Note, you'll need double []â€™s instead of single to index multiple columns. This can be done for all the columns as well!

These methods are great for getting the whole column of data. But what if I just want data from a single subject?

3. indexing data for a single subject

This is also quite easy with pandas. Let's extract the fa data for subject 1_001:

```python
df[['structureID','fa']][df['subjectID'] == '1_001']
```

The first part (df[['structureID','fa']]) indexes the track name and FA columns data. The second part ([df['subjectID'] == '1_001']) is a logical index which says "give me only the data in which the subjectID is 1_001". We now have all the FA tract profile data for subject 1_001! 

Now, to demonstrate the concept even further, let's grab the FA tract profile data for all subjects for the 'anterioFrontalCC' track:

```python
df[['subjectID','fa']][df['structureID'] == 'anterioFrontalCC']
```

I think you get the picture! Play around with some other combinations on your own to really practice!

So we now know how to index columns. The next concept is 'groupby'. 'groupby' is a great function that allows us to group data by a specific column. For example, let's extract the FA tract profile data for each tract for subject 1_001:

```python
df[['structureID','fa']][df['subjectID'] == '1_001'].groupby('structureID')
```

This returns an object, <pandas.core.groupby.generic.DataFrameGroupBy object at 0x7f589783ccc0>. This object we can then compute summary statistics on and will lead us into the next section!

## Summary statistics
Computing summary statistics (i.e. mean, median, minimum, maximum) on pandas dataframes is quite easy! Let's compute the average FA for each tract for subject 1_001 using the mean function:

### Compute mean FA for all tracks in single subject:

```python
df[['structureID','fa']][df['subjectID'] == '1_001'].groupby('structureID').mean()
```

Great! So we have an average FA for each track for subject 1_001.

We can also do this for any other diffusion measure or summary statistic!

## Plotting summary data
The whole point of all of this, and the point of science, is to generate figures. Matplotlib is a great software package for making figures! Let's generate a scatter plot of the average FA and the average ODI data for each track for subject 1_001.

### Make scatter plot of ODI vs FA in one subject across 61 tracks

Let's make a new data structure with just the fa and odi data for subject 1_001:

```python
data = df[['structureID','fa','odi']][df['subjectID'] =='1_001'].groupby('structureID').mean()
data
```

Let's now make a scatter plot of the track averages:

```python
ax = data.plot.scatter(x='fa',y='odi')

# add title
plt.title('1_001 mean FA vs mean ODI track data')
```

As we can see, there appears to be a negative relationship between ODI and FA. This is empirically and mathematically true! Where FA is high, the orientation dispersion (i.e. fanning) of neurites is low (well organized). Congrats! Youâ€™ve replicated an empirical finding in a subject!!!!

##Compute correlation and r2

Letâ€™s identify how correlated the data are:

```python
# compute correlation
corr = data.corr()['odi']['fa']
corr
```

Not a bad correlation! Letâ€™s now fit a linear regression and compute r2 using sklearn:

```python
# import sklearn
from sklearn import linear_model

# initialize linear regression
lm = linear_model.LinearRegression()

# compute linear regression
lm.fit(data['fa'].values.reshape(len(track_names),1),data['odi'].values.reshape(len(track_names),1))

# compute r2
r2 = lm.score(data['fa'].values.reshape(len(track_names),1),data['odi'].values.reshape(len(track_names),1))
r2
```

R2 = 0.656 is quite high! Really strong relationship between fa and odi in this subject! 

### Add regression line to scatter plot and text data

Letâ€™s plot the linear regression!

```python
# plot the scatter plot data again
ax = data.plot.scatter(x='fa',y='odi')

# set title
plt.title('1_001 mean FA vs mean ODI track data')

# plot the regression line
plt.plot(data['fa'].values,lm.predict(data['fa'].values.reshape(len(track_names),1)),color='blue')

# add correlation and r2 to plot
plt.text(0.55, 0.25, 'r = ' + str(corr))
plt.text(0.55, 0.2, 'r2 = ' + str(r2))
```
